```python
def maximos_locales(array):
    maximos = []
    
    for i in range(1, len(array) - 1):
        if array[i] > array[i - 1] and array[i] > array[i + 1]:
            maximos.append(array[i])  
        if array[0] > array[1]:
            maximos.append(array[0])
        if array[-1] > array[-2]:
            maximos.append(array[-1])

    return maximos

final = np.array(datos).flatten()
maximos = maximos_locales(final)

print("MÃ¡ximos locales:", maximos)
```
