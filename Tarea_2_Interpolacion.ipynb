Ejercicio 8 Derivaci√≥n

```python
import numpy as np

h = 0.01
x = np.arange(0.1, 1.2, h)
f_x=np.sqrt(np.tan(x))
f_x_h=np.sqrt(np.tan(x+h))
f_x_2h= np.sqrt(np.tan(x+(2*h)))
derivada_progresiva = (1 / (2 * h)) * ((-3 * f_x) + ((4 * f_x_h)  - f_x_2h))
print(derivada_progresiva)
```

```python
h=0.01
x=np.arange(0.1,1.1,h)
derivada_central=(np.tan(x+h) - np.tan(x-h)) / (2*h)
print(derivada_central)
```

```python
#Estimaci√≥n central
import numpy as np
import matplotlib.pyplot as plt
h=np.logspace(-4, -1, 10)  
x = np.pi/4
funcion=(1/np.cos(x))**2 / (2 * np.sqrt(np.tan(x)))
derivada_central=((1/np.cos(x+h))**2 / (2 * np.sqrt(np.tan(x+h)))-(1/np.cos(x-h))**2 / (2 * np.sqrt(np.tan(x-h))))/(2*h)
derivada_progresiva=((1/np.cos(x+h))**2 / (2 * np.sqrt(np.tan(x+h)))-(1/np.cos(x-h))**2 / (2 * np.sqrt(np.tan(x-h))))/(h)
#Grafica de derivada_central y derivbada_progresiva
fig, (v1, v2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
v1.plot(h, derivada_central, color='black')
v1.set_title('Derivada central')
v1.set_ylabel('Derivada central')
v1.set_xlabel("h")

v2.plot(h, derivada_progresiva, color="red")
v2.set_title("Derivada progresiva")
v2.set_ylabel("Derivada progresiva")
v2.set_xlabel("h")
plt.show()

#Grafica del error nodal para la estimaci√≥n central y progresiva
error_nodal_central= abs(derivada_central-funcion)
error_nodal_progresivo= abs(derivada_progresiva-funcion)

fig, (a1, a2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
a1.plot(error_nodal_central, h, color='black')
a1.set_title('error_nodal_central')
a1.set_ylabel('h')
a1.set_xlabel("error_nodal_central")

a2.plot( error_nodal_progresivo, h, color="red")
a2.set_title("error_nodal_progresivo")
a2.set_ylabel("h")
a2.set_xlabel("erorr_nodal_progresivo")
plt.show()
#Comparaci√≥n
print("Derivada progresiva", derivada_progresiva)
print("Error nodal progresivo",error_nodal_progresivo)
print("Derivada central", derivada_central)
print("Error nodal central",error_nodal_central)

```

Ejercicio 4 Interpolaci√≥n de Lagrange

```python
import pandas as pd
datos = (pd.read_csv('https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/Parabolico.csv'))
print(datos)
```

```python
from scipy.optimize import curve_fit
#Extraer datos del csv
x=datos["X"]
y=datos["Y"]

#INterpolaci√≥n a un polinomiom cuadr√°tico
def polinomio(x, a, b, c):
    return a * x**2 + b * x + c

parametros, _= curve_fit(polinomio, x, y)
a, b, c = parametros

print(f"Coeficientes de la par√°bola: a = {a:.6f}, b = {b:.6f}, c = {c:.6f}")

# Graficar
plt.scatter(x, y)
x_fit = np.linspace(min(x), max(x), 100)
y_fit = polinomio(x_fit, *parametros)
plt.plot(x_fit, y_fit, color="red")
plt.xlabel('Posici√≥n Horizontal (m)')
plt.ylabel('Posici√≥n Vertical (m)')
plt.title('Ajuste de Datos a una Par√°bola')
plt.show()

#Para encontrar el vector, magnitud y angulo, de la velocidad.
g = 9.8  
# Calcular el √°ngulo y la velocidad inicial
theta = np.atan(-b / (2 * a * g))  # √Ångulo en radianes
V0 = np.sqrt(-g / (2 * a * np.cos(theta)**2))  # Velocidad inicial
#Angulos a radianes
theta_deg = np.degrees(theta)

print(f"Velocidad Inicial (V0): {V0:.2f} m/s")
print(f"√Ångulo de Lanzamiento (Œ∏): {theta_deg:.2f}¬∞")
```
