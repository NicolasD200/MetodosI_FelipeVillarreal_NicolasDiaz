```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def valoresarquimedes(a:float, b:float, valor:float):
  return a+b*valor
```

```python
intervalo=np.linspace(0, 2*np.pi)
```

```python
intervalo
```

```python
valores=valoresarquimedes(0,1,intervalo)
valores
```

```python
def polaracartesiano(r:float, teta:float):
  x=r*np.cos(teta)
  y=r*np.sin(teta)
  return (x,y)
```

```python
X=polaracartesiano(valores, intervalo)[0]
Y=polaracartesiano(valores, intervalo)[1]
```

```python
plt.plot(X,Y)
```
