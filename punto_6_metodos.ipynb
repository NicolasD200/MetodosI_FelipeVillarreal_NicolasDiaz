```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la discretizaci칩n en los ejes x e y
X = np.linspace(-4, 4, 25)
Y = np.linspace(-4, 4, 25)
x, y = np.meshgrid(X, Y, indexing='ij')

# Definir la funci칩n potencial del flujo
def potencialdeflujo(x, y, V=2, R=2):
    # Evitar dividir por cero; usamos un peque침o epsilon donde x^2 + y^2 = 0
    epsilon = 1e-10
    denominador = np.maximum(x**2 + y**2, epsilon)  # Corregir a x**2 + y**2
    return V * x * (1 - (R**2 / denominador))

# Definimos las derivadas centrales
def CentralDerivativeX(f, x, y, h=0.001):
    return (f(x + h, y) - f(x - h, y)) / (2 * h)

def CentralDerivativeY(f, x, y, h=0.001):
    return -(f(x, y + h) - f(x, y - h)) / (2 * h)

# Calculamos el potencial de flujo para la malla de puntos
flujo = potencialdeflujo(x, y)

# Calculamos las derivadas centrales para cada punto en la malla
velocidadx = np.zeros_like(x)
velocidady = np.zeros_like(y)

R = 2  # Radio del cilindro
for i in range(len(X)):
    for j in range(len(Y)):
        if x[i, j]**2 + y[i, j]**2 <= R**2:  # Dentro del cilindro, velocidad = 0
            velocidadx[i, j] = 0
            velocidady[i, j] = 0
        else:
            velocidadx[i, j] = CentralDerivativeX(potencialdeflujo, x[i, j], y[i, j])
            velocidady[i, j] = CentralDerivativeY(potencialdeflujo, x[i, j], y[i, j])

# Dibujamos el campo de velocidades
plt.figure(figsize=(6, 6))
plt.quiver(X, Y, velocidadx, velocidady)
plt.title("Campo de velocidades cerca de un cilindro s칩lido")
plt.xlabel('x [cm]')
plt.ylabel('y [cm]')
plt.axis('equal')
plt.show()
```
